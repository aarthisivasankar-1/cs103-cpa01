# Using Pivot Tables and Plots to Analyze Data

To answer both questions stated at the very beginning of the notebook, we can utilize both pivot tables and plots to closely analyze the data.

For the first question, I am curious to see whether the average salary of jobs had an effect in what rating was given for each job.

Firstly, I want to print an array of all the column names from this dataset so that I know which data types I can reference and get access to:

```python
list(df.columns)
```

Similarly, I want to print the index array using numpy to better organize each job title with its respective information:

```python
df.to_numpy()
```

For my first question, because I want to examine the effect average salary has had on the ratings given to each job title, I want to extract only the average salary and ratings data to then plot and get a better understanding as to whether they have any sort of association/correlationn.

Printing Avg Salary Data:

```python
df['Avg Salary(K)']
```

Printing Rating Data:

```python
df['Rating']
```

Now, we can make a plot with Average Salaries by Ratings to analyze whether there are any correlations between both fields of data:

```python
plt.figure(figsize=(25,15))
plt.scatter(df['Avg Salary(K)'],df['Rating'])
plt.xlabel("Avg Salary (in Thousands)")
plt.ylabel("Rating")
plt.grid()
```

This scatter plot seems to have dispersed data points, not clearly depicting any positive or negative trendlines. Despite having a few major outliers, it seems that most of the ratings of all the data science jobs are at least greater than 3.0, which is a good sign! I will discuss more of my analyses at the end in the discussion cell.

Let's say I want to further examine the realtion between ratings and salaries for data science roles. Let's use a pivot table to see whether roles with higher ratings tend to have higher average salaries or lower, or show no association at all. Note that some values say 0.0 because there were no average salaries reported for thise specific roles with the ratings provided. 

```python
df2 = pd.pivot_table(df,values="Avg Salary(K)",index=["Job Title"], columns=["Rating"], dropna="False", fill_value=0)
df2
```

```python
df2.plot(figsize=(30,10))
plt.grid()
plt.xlabel("Data Science Roles")
plt.ylabel("Average Salary (in Thousands)")
```

This plot shows the ratings of different data science roles with its respective average paying salary. Some noticeable points to note are a Data Enterprise Architect role at a company that pays an average salary of approximately slightly over 250K has an average rating of 2.3, while the Senior Data Scientist Artificial Intelligence role that pays an average salary of approximately 240K has an average rating of 4.3. (Note that the legend reports similar colors for the rating legends, so I had to analyze the plot specifically along with the dataframe table to confirm the rating values I analyzed). This shows how just because a role is paying more, does not automatically mean it will be rated higher. Clearly, there are more factors than just salary that go into how a specific role at a company is rated as.

We can also utilize the pandas groupby function to help us group the data from the data set by a particular value to better understand the data more. For example,

```python
gb = df.groupby("Rating")
gb.first()
```

Using the groupby function, I have printed out the first element of each group based on their ranking. This allows us to see the roles with the best/worst ranking. For example, seems as though a Principal Data Scientist role having the highest salary estimate of approximately 200K-250K has the worst rating of them all! Really goes to show how salary is not the only factor that employees value in their job career roles.
